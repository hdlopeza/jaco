```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
file = '.\example.jpeg'
```

```python
a = tf.io.read_file(file)
```

```python
b = tf.io.serialize_tensor(a)
```

```python
c = b.numpy()
```

```python
d = tf.io.parse_tensor(c, tf.string)
```

```python
e = tf.image.decode_jpeg(d)
```

```python
plt.imshow(e)
```
